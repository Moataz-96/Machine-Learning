```python
def get_df_record(i):
    tagline_record = df['tagline'].iloc[i] if df['tagline'].notnull()[i+1] else "none"
    runtime_record = df['runtime'].iloc[i] if df['runtime'].notnull()[i+1] else 0
    original_title_record = df['original_title'].iloc[i] if df['original_title'].notnull()[i+1] else "none"
    original_language_record = df['original_language'].iloc[i] if df['original_language'].notnull()[i+1] else "none"
    popularity_record = df['popularity'].iloc[i] if df['popularity'].notnull()[i+1] else 0
    budget_record = df['budget'].iloc[i] if df['budget'].notnull()[i+1] else 0
    status_record = df['status'].iloc[i] if df['status'].notnull()[i+1] else "none"
    revenue_record = df['revenue'].iloc[i] if df['revenue'].notnull()[i+1] else 0
    
    genres_record = " ".join([genre['name'] for genre in ast.literal_eval(df['genres'].iloc[i])]) if df['genres'].notnull()[i+1] else "none"
    
    production_companies_record = " ".join([production_companies['name'].replace(" ","-") for production_companies in ast.literal_eval(df['production_companies'].iloc[i])]) if df['production_companies'].notnull()[i+1] else "none" 
    
    production_countries_record = " ".join([production_countries['name'].replace(" ","-") for production_countries in ast.literal_eval(df['production_countries'].iloc[i])]) if df['production_countries'].notnull()[i+1] else "none"
    
    spoken_languages_record = " ".join([spoken_languages['name'].replace(" ","-") for spoken_languages in ast.literal_eval(df['spoken_languages'].iloc[i])]) if df['spoken_languages'].notnull()[i+1] else "none"
    
    Keywords_record = " ".join([Keywords['name'].replace(" ","-") for Keywords in ast.literal_eval(df['Keywords'].iloc[i])]) if df['Keywords'].notnull()[i+1] else "none"
    
    cast_record = " ".join([cast['name'].replace(" ","-") for cast in ast.literal_eval(df['cast'].iloc[i])]) if df['cast'].notnull()[i+1] else "none"
    
    crew_record = " ".join([crew['name'].replace(" ","-") for crew in ast.literal_eval(df['crew'].iloc[i])]) if df['crew'].notnull()[i+1] else "none"
    
    belongs_to_collection_record = " ".join([belongs_to_collection['name'].replace(" ","-") for belongs_to_collection in ast.literal_eval(df['belongs_to_collection'].iloc[i])]) if df['belongs_to_collection'].notnull()[i+1] else "none"
    
    return {"tagline_record":tagline_record,
           "runtime_record":runtime_record,
           "original_title_record":original_title_record,
           "original_language_record":original_language_record,
           "popularity_record":popularity_record,
           "budget_record":budget_record,
           "status_record":status_record,
           "genres_record":genres_record,
           "production_companies_record":production_companies_record,
           "production_countries_record":production_countries_record,
           "spoken_languages_record":spoken_languages_record,
           "Keywords_record":Keywords_record,
           "cast_record":cast_record,
           "crew_record":crew_record,
           "belongs_to_collection_record":belongs_to_collection_record,
           "revenue_record":revenue_record
           }
```

```python
from tensorflow.train import FloatList,BytesList,Int64List
from tensorflow.train import Features, Feature, Example

def TFRecordWriter(file_path="my_data.tfrecords"):
    with tf.io.TFRecordWriter(file_path) as f:
        for i in range(df.shape[0]):
            records = get_df_record(df,i)
            example = Example(features=Features(feature = {
                "popularity_record":Feature(float_list=FloatList(value=[records['popularity_record']])),
                "runtime_record":Feature(float_list=FloatList(value=[records['runtime_record']])),
                "budget_record":Feature(int64_list=Int64List(value=[records['budget_record']])),
                "tagline_record":Feature(bytes_list=BytesList(value=[records['tagline_record'].encode("utf-8")])), 
                "original_title_record":Feature(bytes_list=BytesList(value=[records['original_title_record'].encode("utf-8")])),
                "original_language_record":Feature(bytes_list=BytesList(value=[records['original_language_record'].encode("utf-8")])),
                "status_record":Feature(bytes_list=BytesList(value=[records['status_record'].encode("utf-8")])), 
                "genres_record":Feature(bytes_list=BytesList(value=[records['genres_record'].encode("utf-8")])),
                "production_companies_record":Feature(bytes_list=BytesList(value=[records['production_companies_record'].encode("utf-8")])),
                "production_countries_record":Feature(bytes_list=BytesList(value=[records['production_countries_record'].encode("utf-8")])),
                "spoken_languages_record":Feature(bytes_list=BytesList(value=[records['spoken_languages_record'].encode("utf-8")])),
                "Keywords_record":Feature(bytes_list=BytesList(value=[records['Keywords_record'].encode("utf-8")])),
                "cast_record":Feature(bytes_list=BytesList(value=[records['cast_record'].encode("utf-8")])),
                "crew_record":Feature(bytes_list=BytesList(value=[records['crew_record'].encode("utf-8")])),
                "belongs_to_collection_record":Feature(bytes_list=BytesList(value=[records['belongs_to_collection_record'].encode("utf-8")])),
                "revenue_record":Feature(float_list=FloatList(value=[records['revenue_record']])),
            }))
            f.write(example.SerializeToString())
        
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(np.array(df[["popularity","runtime","budget"]]))
X_mean = scaler.mean_
X_std = scaler.scale_
```

```python
X_mean[1]
```

```python
popularity_record = tf.feature_column.numeric_column("popularity_record",
                                                     normalizer_fn=lambda x:(x-X_mean[0])/X_std[0])
runtime_record = tf.feature_column.numeric_column("runtime_record",
                                                     normalizer_fn=lambda x:(x-X_mean[1])/X_std[1])
budget_record = tf.feature_column.numeric_column("budget_record",
                                                     normalizer_fn=lambda x:(x-X_mean[2])/X_std[2],dtype=tf.dtypes.int64)

tagline_record = tf.feature_column.categorical_column_with_hash_bucket("tagline_record", 10)
tagline_record = tf.feature_column.embedding_column(tagline_record,dimension=20)

original_title_record = tf.feature_column.categorical_column_with_hash_bucket("original_title_record", 10000)
original_title_record = tf.feature_column.embedding_column(original_title_record,dimension=20)

original_language_record = tf.feature_column.categorical_column_with_hash_bucket("original_language_record", 10000)
original_language_record = tf.feature_column.embedding_column(original_language_record,dimension=20)

production_companies_record = tf.feature_column.categorical_column_with_hash_bucket("production_companies_record", 10000)
production_companies_record = tf.feature_column.embedding_column(production_companies_record,dimension=20)

spoken_languages_record = tf.feature_column.categorical_column_with_hash_bucket("spoken_languages_record", 10000)
spoken_languages_record = tf.feature_column.embedding_column(spoken_languages_record,dimension=20)

Keywords_record = tf.feature_column.categorical_column_with_hash_bucket("Keywords_record", 10000)
Keywords_record = tf.feature_column.embedding_column(Keywords_record,dimension=20)

cast_record = tf.feature_column.categorical_column_with_hash_bucket("cast_record", 10000)
cast_record = tf.feature_column.embedding_column(cast_record,dimension=20)

crew_record = tf.feature_column.categorical_column_with_hash_bucket("crew_record", 10000)
crew_record = tf.feature_column.embedding_column(crew_record,dimension=20)

belongs_to_collection_record = tf.feature_column.categorical_column_with_hash_bucket("belongs_to_collection_record", 10000)
belongs_to_collection_record = tf.feature_column.embedding_column(belongs_to_collection_record,dimension=20)

genres_record = tf.feature_column.categorical_column_with_hash_bucket("genres_record", 10000)
genres_record = tf.feature_column.embedding_column(genres_record,dimension=20)

status_record = tf.feature_column.categorical_column_with_hash_bucket("status_record", 2)
status_record = tf.feature_column.indicator_column(status_record)

production_countries_record = tf.feature_column.categorical_column_with_hash_bucket("production_countries_record", 10000)
production_countries_record = tf.feature_column.embedding_column(production_countries_record,dimension=20)

revenue_record = tf.feature_column.numeric_column("revenue_record")
```

```python
columns = [popularity_record,runtime_record,budget_record,tagline_record,original_title_record,
          original_language_record,production_companies_record,spoken_languages_record,Keywords_record,
          cast_record,crew_record,belongs_to_collection_record,genres_record,status_record,
          production_countries_record,revenue_record]
```

```python
feature_descriptions = tf.feature_column.make_parse_example_spec(columns)
feature_descriptions
```

```python
def parse_example(serialized_examples):
    example = tf.io.parse_example(serialized_examples,feature_descriptions)
    targets = example.pop("revenue_record")
    return example,targets
batch_size = 512
dataset = tf.data.TFRecordDataset(["my_data.tfrecords"])
dataset = dataset.repeat().shuffle(10000).batch(batch_size).map(parse_example)
```

```python
test_dataset = dataset.take(300) 
train_dataset = dataset.skip(300)
valid_dataset = test_dataset.take(100)
```

```python
for x in train_dataset.take(1):
    print(x)
```

```python
batch_size=512
columns_without_target = columns[:-1]
model = keras.models.Sequential([
    keras.layers.DenseFeatures(feature_columns=columns_without_target),
    keras.layers.Dense(1)
])
model.compile(loss="mse",optimizer=keras.optimizers.SGD(lr=0.1,clipnorm=1),metrics=[tf.keras.metrics.MeanSquaredLogarithmicError()])
```

```python
model.fit(train_dataset, steps_per_epoch= 2700// batch_size, epochs=10000,validation_data=valid_dataset)
```

```python
# keras.models.save_model(model,"imdb_box_office.h5")
model.save_weights('weights/imdb_box_office')
```
